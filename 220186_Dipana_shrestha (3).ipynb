```python
import pandas as pd

# 3 Tasks - To-Do:

# 3.1
# Task 1: Introduction to Data Exploration - Initial Exercises:

# 1. Data Exploration and Understanding:

# Dataset Overview:
# 1. Load the dataset and display the first 10 entries.
file_path = "/content/drive/MyDrive/WHR-2024-5CS037.csv"
df = pd.read_csv(file_path)
print("\nDisplaying the first 10 entries of the dataset:")
print(df.head(10))

# 2. Determine the total number of rows and columns in the dataset.
print("\nTotal number of rows and columns:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# 3. List all columns and their respective data types.
print("\nList of columns and their data types:")
print(df.dtypes)
```

```python
# Basic Statistics:
# 1. Compute the mean, median, and standard deviation for the 'score' column.
mean_score = df['score'].mean()
median_score = df['score'].median()
std_score = df['score'].std()

print("Statistical measures for the 'score' column:")
print(f"Mean: {mean_score}")
print(f"Median: {median_score}")
print(f"Standard Deviation: {std_score}")

# 2. Find the country with the highest and lowest happiness scores.
country_with_highest_score = df.loc[df['score'].idxmax(), 'Country name']
country_with_lowest_score = df.loc[df['score'].idxmin(), 'Country name']

print("\nCountries with the highest and lowest happiness scores:")
print(f"Country with Highest Score: {country_with_highest_score}")
print(f"Country with Lowest Score: {country_with_lowest_score}")
```

```python
# Missing Values:
# 1. Verify if there are any missing values in the dataset and display the count of missing values per column.
print("Count of Missing Values in the Dataset:")
missing_values_count = df.isnull().sum()

print(missing_values_count)
```

```python
# Filtering and Sorting:
# 1. Filter the dataset to display only countries with a Score above 7.5.
filtered_data = df[df['score'] > 7.5]

print("Filtered Data (Score > 7.5):")
print(filtered_data)

# 2. Sort the filtered data by GDP per Capita in descending order and show the top 10 rows.
sorted_data = filtered_data.sort_values(by='Log GDP per capita', ascending=False)

print("\nTop 10 countries sorted by GDP per Capita in descending order:")
print(sorted_data.head(10))
```

```python
# Adding New Columns:
# 1. Add a new column named 'Happiness Category' that classifies countries into three categories based on their Score:
#    Low − (Score < 4)
#    Medium − (4 ≤ Score ≤ 6)
#    High − (Score > 6)

def categorize_happiness(score):
    if score < 4:
        return "Low"
    elif 4 <= score <= 6:
        return "Medium"
    else:
        return "High"

df['Happiness Category'] = df['score'].apply(categorize_happiness)

print("\nDataset with the Happiness Category:")
print(df[['Country name', 'score', 'Happiness Category']].head())
```

```python
# Assuming df is already loaded, if not, load the dataset as follows:
# df = pd.read_csv('path_to_dataset.csv')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar Plot: Plot the top 10 happiest countries by Score using a bar chart.
top_10_happiest = df.sort_values(by='score', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='score', y='Country name', data=top_10_happiest, color='skyblue')
plt.title('Top 10 Happiest Countries by Score')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()

# 2. Line Plot: Plot the top 10 unhappiest countries by Score using a line chart.
top_10_unhappiest = df.sort_values(by='score', ascending=True).head(10)

plt.figure(figsize=(10, 6))
sns.lineplot(x='Country name', y='score', data=top_10_unhappiest, marker='o', color='red')
plt.title('Top 10 Unhappiest Countries by Score')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()

# 3. Plot a histogram for the Score column to show its distribution and also interpret.
plt.figure(figsize=(10, 6))
sns.histplot(df['score'], bins=20, kde=True, color='blue')
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()

# 4. Scatter Plot: Plot a scatter plot between GDP per Capita and Score to visualize their relationship.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=df, color='green')
plt.title('Scatter Plot between GDP per Capita and Happiness Score')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# 3.2 Problem - 2 - Advanced Data Exploration Task:

# Task - 1 - Setup Task - Preparing the South-Asia Dataset:

# Step 1: Define a list of countries in South Asia
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                         "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Step 2: Filter the dataset to include only the countries from South Asia
south_asian_data = df[df['Country name'].isin(south_asian_countries)]

# Step 3: Save the filtered dataset as a new CSV file
south_asian_data.to_csv('south_asian_happiness_data.csv', index=False)

# Display the first 5 rows of the filtered dataset for verification
print(south_asian_data.head())
```

```python
# Task - 2 - Composite Score Ranking

# 1. Create a new column in the South Asia DataFrame called 'Composite Score' that combines the following metrics:
# Composite Score = 0.40 × GDP per Capita + 0.30 × Social Support + 0.30 × Healthy Life Expectancy
south_asian_data.loc[:, 'Composite Score'] = (0.40 * south_asian_data['Log GDP per capita'] +
                                              0.30 * south_asian_data['Social support'] +
                                              0.30 * south_asian_data['Healthy life expectancy'])

# 2. Sort the South Asia DataFrame in descending order based on the Composite Score
south_asian_data = south_asian_data.sort_values(by='Composite Score', ascending=False)

# 3. Visualize the top 5 countries based on Composite Score using a horizontal bar chart
top_5_composite = south_asian_data.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x='Composite Score', y='Country name', data=top_5_composite, color='skyblue')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.xlabel('Composite Score')
plt.ylabel('Country')
plt.show()

# 4. Compare whether the rankings based on the Composite Score align with the original Score and discuss the findings using a visualization
# We will visualize the comparison between the original score and the composite score for the top 5 countries
plt.figure(figsize=(10, 6))
sns.barplot(x='Composite Score', y='Country name', data=top_5_composite, color='blue', label='Composite Score')
sns.barplot(x='score', y='Country name', data=top_5_composite, color='orange', label='Original Score')
plt.title('Comparison of Composite Score vs Original Score for the Top 5 Countries')
plt.xlabel('Score')
plt.ylabel('Country')
plt.legend()
plt.show()
```

```python
# Task - 3 - Outlier Detection

# 1. Detect outlier countries in South Asia based on Score and GDP per Capita
Q1_score = south_asian_data['score'].quantile(0.25)
Q3_score = south_asian_data['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asian_data['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asian_data['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# 2. Identify outliers using the 1.5 × IQR rule
score_outliers = south_asian_data[(south_asian_data['score'] < (Q1_score - 1.5 * IQR_score)) |
                                  (south_asian_data['score'] > (Q3_score + 1.5 * IQR_score))]

gdp_outliers = south_asian_data[(south_asian_data['Log GDP per capita'] < (Q1_gdp - 1.5 * IQR_gdp)) |
                                 (south_asian_data['Log GDP per capita'] > (Q3_gdp + 1.5 * IQR_gdp))]

outliers = pd.concat([score_outliers, gdp_outliers]).drop_duplicates()

# 3. Plot a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asian_data, color='blue', label='Non-Outliers')
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers, color='red', label='Outliers')
plt.title('Scatter Plot: GDP per Capita vs Score with Outliers Highlighted')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
# Task - 4 - Exploring Trends Across Metrics

# 1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation(Pearson correlation)
# with the Score for South Asian countries.
freedom_corr = south_asian_data['Freedom to make life choices'].corr(south_asian_data['score'])
generosity_corr = south_asian_data['Generosity'].corr(south_asian_data['score'])

print(f"Pearson correlation between Freedom to Make Life Choices and Score: {freedom_corr}")
print(f"Pearson correlation between Generosity and Score: {generosity_corr}")

# 2. Create scatter plots with trendlines for these metrics against the Score.
plt.figure(figsize=(10, 6))
sns.regplot(x='Freedom to make life choices', y='score', data=south_asian_data, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Scatter Plot: Freedom to Make Life Choices vs Score')
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Score')
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(x='Generosity', y='score', data=south_asian_data, scatter_kws={'color':'green'}, line_kws={'color':'red'})
plt.title('Scatter Plot: Generosity vs Score')
plt.xlabel('Generosity')
plt.ylabel('Score')
plt.show()
```

```python
# Task - 5 - Gap Analysis

# 1. Create a new column, 'GDP-Score Gap', which calculates the difference between GDP per Capita and the Happiness Score for each South Asian country.
south_asian_data['GDP-Score Gap'] = south_asian_data['Log GDP per capita'] - south_asian_data['score']

# 2. Rank the South Asian countries based on the GDP-Score Gap in both ascending and descending order.
# Sorting by ascending gap
ascending_gap = south_asian_data[['Country name', 'GDP-Score Gap']].sort_values(by='GDP-Score Gap', ascending=True)

# Sorting by descending gap
descending_gap = south_asian_data[['Country name', 'GDP-Score Gap']].sort_values(by='GDP-Score Gap', ascending=False)

# 3. Create bar charts to highlight the top 3 countries with the highest positive and negative gaps.
top_3_positive_gap = descending_gap.head(3)
top_3_negative_gap = ascending_gap.head(3)

# Bar plot for countries with the largest positive GDP-Score gap
plt.figure(figsize=(10, 6))
sns.barplot(x='GDP-Score Gap', y='Country name', data=top_3_positive_gap)
plt.title('Top 3 Countries with the Highest Positive GDP-Score Gap')
plt.xlabel('GDP-Score Gap')
plt.ylabel('Country')
plt.show()

# Bar plot for countries with the largest negative GDP-Score gap
plt.figure(figsize=(10, 6))
sns.barplot(x='GDP-Score Gap', y='Country name', data=top_3_negative_gap)
plt.title('Top 3 Countries with the Highest Negative GDP-Score Gap')
plt.xlabel('GDP-Score Gap')
plt.ylabel('Country')
plt.show()
```

```python
# Task - 1 - Preparing the Middle Eastern Dataset

# Define the list of Middle Eastern countries
middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

# Filter the dataframe to include only Middle Eastern countries
middle_east_df = df[df['Country name'].isin(middle_east_countries)]

# Save the filtered data to a new CSV file
middle_east_df.to_csv("middle_east_data.csv", index=False)

# Display the first few rows of the filtered data for verification
print(middle_east_df.head())
```

```python
# 1. Descriptive Statistics:

# • Calculate the mean and standard deviation of the happiness scores for South Asia and the Middle East.
south_asia_mean = south_asian_data['score'].mean()
south_asia_std = south_asian_data['score'].std()

middle_east_mean = middle_east_df['score'].mean()
middle_east_std = middle_east_df['score'].std()

print(f"South Asia - Average: {south_asia_mean}, Standard Deviation: {south_asia_std}")
print(f"Middle East - Average: {middle_east_mean}, Standard Deviation: {middle_east_std}")

# • Which region has a higher average happiness score?
if south_asia_mean > middle_east_mean:
    print("South Asia has a higher average happiness score.")
else:
    print("Middle East has a higher average happiness score.")
```

```python
# 2. Top and Bottom Performers:

# • Identify the top 3 and bottom 3 countries in each region based on their happiness score.
south_asia_top_3 = south_asian_data[['Country name', 'score']].sort_values(by='score', ascending=False).head(3)
south_asia_bottom_3 = south_asian_data[['Country name', 'score']].sort_values(by='score').head(3)

middle_east_top_3 = middle_east_df[['Country name', 'score']].sort_values(by='score', ascending=False).head(3)
middle_east_bottom_3 = middle_east_df[['Country name', 'score']].sort_values(by='score').head(3)

# • Create bar charts to compare these top and bottom performers.
plt.figure(figsize=(12, 6))

# Plot for South Asia
plt.subplot(1, 2, 1)
sns.barplot(x='score', y='Country name', data=south_asia_top_3, hue='Country name', palette='Blues', legend=False)
plt.title('Top 3 Countries in South Asia by Happiness Score')

plt.subplot(1, 2, 2)
sns.barplot(x='score', y='Country name', data=south_asia_bottom_3, hue='Country name', palette='Reds', legend=False)
plt.title('Bottom 3 Countries in South Asia by Happiness Score')

plt.tight_layout()
plt.show()

# Plot for Middle East
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.barplot(x='score', y='Country name', data=middle_east_top_3, hue='Country name', palette='Blues', legend=False)
plt.title('Top 3 Countries in Middle East by Happiness Score')

plt.subplot(1, 2, 2)
sns.barplot(x='score', y='Country name', data=middle_east_bottom_3, hue='Country name', palette='Reds', legend=False)
plt.title('Bottom 3 Countries in Middle East by Happiness Score')

plt.tight_layout()
plt.show()
```

```python
# 3. Metric Comparisons:

# • Use grouped bar charts to compare key indicators such as GDP per Capita, Social Support, and Healthy Life Expectancy across South Asia and the Middle East.
metrics = ['Log GDP per capita', 'Social support', 'Healthy life expectancy']

south_asia_metrics = south_asian_data[['Country name'] + metrics].melt(id_vars='Country name', value_vars=metrics, var_name='Metric', value_name='Value')
middle_east_metrics = middle_east_df[['Country name'] + metrics].melt(id_vars='Country name', value_vars=metrics, var_name='Metric', value_name='Value')

south_asia_metrics['Region'] = 'South Asia'
middle_east_metrics['Region'] = 'Middle East'

comparison_df = pd.concat([south_asia_metrics, middle_east_metrics]).reset_index(drop=True)

plt.figure(figsize=(12, 6))
sns.barplot(x='Metric', y='Value', hue='Region', data=comparison_df, palette='Set2')

plt.title('Comparison of Key Metrics Between South Asia and Middle East')
plt.xlabel('Metric')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```

```python
# 4. Happiness Disparity:

# • Calculate the range (difference between max and min) and the coefficient of variation (CV) for the happiness scores in both regions.
south_asia_score_range = south_asian_data['score'].max() - south_asian_data['score'].min()
middle_east_score_range = middle_east_df['score'].max() - middle_east_df['score'].min()

south_asia_cv = south_asian_data['score'].std() / south_asian_data['score'].mean()
middle_east_cv = middle_east_df['score'].std() / middle_east_df['score'].mean()

print(f"Happiness Score Range in South Asia: {south_asia_score_range}")
print(f"Happiness Score Range in Middle East: {middle_east_score_range}")
print(f"Coefficient of Variation (CV) in South Asia: {south_asia_cv}")
print(f"Coefficient of Variation (CV) in Middle East: {middle_east_cv}")
```
